```python
import os
import pandas as pd
import json
import ast
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
data_path = "../../legal"
train_df = pd.read_csv(os.path.join(data_path, "train.csv"))
```

```python
train_df['context'] = train_df['context'].apply(ast.literal_eval)
```

```python
train_df['cid'] = train_df['cid'].apply(lambda x: list(map(int, x.strip('[]').split())))
```

```python
texts = []
labels = []
questions = []
for i, row in tqdm(train_df.iterrows(), total=train_df.shape[0]):
    question = row["question"]
    contexts = row["context"]
    for context in contexts:
        questions.append(question)
        texts.append(context)
        labels.append(1)

df = pd.DataFrame()
df["text"] = texts
df["question"] = questions
df["label"] = labels
```

```python
df[:20]
```

```python
df.text = df.text.apply(lambda x: x.replace("\r","\n"))
```

```python
df.to_csv("../processed/train_stage1_ranking.csv",index=False)
```
